```python
import numpy as np
import triangle
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay
from numpy import genfromtxt
```

```python
points = genfromtxt('data/Buff_noHole.csv', delimiter=',')
```

```python
plt.scatter(points[:, 0], points[:, 1])
plt.show()
```

```python
tri= Delaunay(points)
```

```python
plt.triplot(points[:, 0], points[:, 1], tri.simplices.copy())
plt.plot(points[:, 0], points[:, 1], 'o')
plt.show()
```

```python
points = genfromtxt('data/Buff_withHoles.csv', delimiter=',')
tri= Delaunay(points)
plt.triplot(points[:, 0], points[:, 1], tri.simplices.copy())
plt.plot(points[:, 0], points[:, 1], 'o')
plt.show()
```
